```python
import opendatasets as od
import pandas as pd
od.download('https://www.kaggle.com/datasets/arianazmoudeh/airbnbopendata')
```

```python
df = pd.read_csv('/content/airbnbopendata/Airbnb_Open_Data.csv')
data = pd.DataFrame(df)
data.head()
```

```python
#to know the number of columns of the data and know thier names in large datasets
data.columns

```

**The main steps to follow when we get a data set**
1. Deleting unwanted columns firstly after identyfing it using info() method
2. Renaming the columns if needed
3. Droping Duplicates
4. cleaning individual columns
5. Removing NAN values from the datasets
6. cheacking for other transformation like normalization and encoding the catogarical datas

**Deleting unwanted columns**

```python
data.info()
```

```python
data.columns
```

```python
# removing the columns having high missing values
data_drop = [ 'reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license','id']
data.drop(data_drop, axis=True,inplace=True )
```

```python
print(data.shape)
data.columns
```

**Renaming the columns if needed **

```python
all_columns = data.columns
renamed_columns = []
for i in all_columns:
  renamed_columns.append(i.lower())
renamed_columns
```

```python
data.columns = renamed_columns
data.columns
```

**Checking the Duplicates and removing it.....**

```python
data.duplicated()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

**Removing the NaN values from the dataset**

```python
data.isnull().sum()
```

```python
data.isnull()#isna
```

```python
data.isna().sum()
```

```python
data.isna()
```

```python
data.dropna(inplace=True)
```
